### Visualisation

```python
mpl.__version__
```

```python
print(plt.style.available)
mpl.style.use(['ggplot'])
```

```python
uzb = df_can.loc['Uzbekistan', years]
uzb.head()
```

```python
uzb.plot()
```

```python
uzb.index = uzb.index.map(int)
uzb.plot(kind='line')

plt.title('Immigration to Canada from Uzbekistan')
plt.xlabel('Years')
plt.ylabel('Number of people')


plt.show()
```

```python
haiti = df_can.loc['Haiti', years]
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigration to Canada from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of people')

plt.text(2000, 6000, '2010 Earthquake')

plt.show()
```
